```python
import numpy as np

# Función para leer las matrices desde el archivo
def read_digit_matrices(filename):
    with open(filename, 'r') as file:
        content = file.readlines()
    
    # Convertir el contenido en matrices de 7x5
    matrices = []
    for i in range(0, len(content), 7):  # Cada matriz es de 7 líneas
        matrix = []
        for j in range(7):
            # Convertir cada línea en una lista de enteros
            line = list(map(int, content[i + j].strip().split()))
            matrix.append(line)
        matrices.append(np.array(matrix))
    
    return matrices

# Función para generar variaciones con ruido gaussiano
def generate_variations(matrix, alpha, num_variations=10):
    variations = []
    
    for _ in range(num_variations):
        # Generar ruido gaussiano con media 0 y desviación estándar alpha
        noise = np.random.normal(0, alpha, matrix.shape)
        
        # Agregar ruido a la matriz original
        noisy_matrix = matrix + noise
        
        # Asegurarse de que los valores siguen siendo binarios (0 o 1)
        noisy_matrix = np.where(noisy_matrix > 0.5, 1, 0)
        
        variations.append(noisy_matrix)
    
    return variations

# Parámetro de ruido (alpha)
alpha = 0.2

# Leer las matrices del archivo
filename = '../inputs/TP3-ej3-digitos.txt'
digit_matrices = read_digit_matrices(filename)

# Nombre del archivo de salida
output_filename = f'test_noise_{alpha}.txt'

# Guardar todas las variaciones en un archivo
with open(output_filename, 'w') as output_file:
    for index, digit_matrix in enumerate(digit_matrices):
        variations = generate_variations(digit_matrix, alpha)
        
        for i, var in enumerate(variations):
            for row in var:
                output_file.write(' '.join(map(str, row)) + '\n')
print(f"Variaciones guardadas en {output_filename}")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Función para leer los datos del archivo y procesarlos
def procesar_datos(file_path):
    with open(file_path, 'r') as f:
        # Leer todas las líneas del archivo
        lines = f.readlines()
    
    # Inicializar lista para almacenar matrices de dígitos
    digitos = {i: [] for i in range(10)}
    
    # Asumir que las representaciones de los dígitos están en bloques de 7x5
    digit_matrix = []
    digit_count = 0
    
    for line in lines:
        if len(line.strip()) > 0:
            # Convertir la línea en una lista de enteros
            digit_matrix.append([int(x) for x in line.strip().split()])
            
            # Si tenemos una matriz de 7x5, almacenarla en el dígito correspondiente
            if len(digit_matrix) == 7:
                digito_actual = digit_count // 100  # Suponer que hay 100 representaciones por dígito
                digitos[digito_actual].append(np.array(digit_matrix))
                digit_matrix = []  # Reiniciar la matriz temporal
                digit_count += 1

    return digitos

# Función para generar el heatmap para cada dígito usando seaborn
def generar_heatmaps(digitos):
    for digito, matrices in digitos.items():
        # Sumar todas las representaciones de un dígito
        suma_matriz = sum(matrices)
        
        # Graficar el heatmap usando seaborn
        plt.figure(figsize=(4, 3))
        sns.heatmap(suma_matriz, cmap='gray_r', cbar=True, square=True, xticklabels=False, yticklabels=False)
        plt.title(f'Heatmap del dígito {digito}')
        plt.show()

# Ruta al archivo
file_path = 'test_noise_0.3.txt'

# Procesar los datos y generar los heatmaps
digitos = procesar_datos(file_path)
generar_heatmaps(digitos)
```
